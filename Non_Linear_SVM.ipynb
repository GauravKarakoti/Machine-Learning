```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.svm import SVC
```

```python
# Step 1: Create a non-linear dataset
X, y = make_circles(
    n_samples=100,
    noise=0.1,
    factor=0.3,
    random_state=42
)
```

```python
# Step 2: Train Non-Linear SVM (RBF Kernel)
model = SVC(kernel='rbf')
model.fit(X, y)
```

```python
# Step 3: Predict a new point
new_point = np.array([[0.0, 0.0]])
prediction = model.predict(new_point)
print("Predicted Class: ", prediction[0])
```

```python
# Step 4: Create Decision Boundary
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500),
)
z = model.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
```

```python
# Step 5: Plot Decision Boundary
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, z, alpha=0.3)
plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    edgecolors='k',
    s=60,
    label='Training Data'
)
# Step 6: Plot New Point
plt.scatter(
    new_point[:,0],
    new_point[:,1],
    color='red',
    s=200,
    marker='*',
    label='New Point'
)
plt.title("Non-Linear SVM (RBF Kernel)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
