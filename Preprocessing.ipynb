```python
# ===============================
# IMPORTING LIBRARIES
# ===============================
import pandas as pd
import numpy as np
```

```python
# ===============================
# 1. LOAD THE DATA
# ===============================
df = pd.read_csv("Messy_Employee_dataset.csv")
print("Original Shape: ", df.shape)
print("\nFirst 5 rows:")
print(df.head())
```

```python
# ===============================
# 2. REMOVE DUPLICATES
# ===============================
print("\n--- 2. Removing Duplicates ---")
print("Duplicate rows found: ", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("New Shape after removing duplicates: ", df.shape)
```

```python
# ===============================
# 3. HANDLE NOISE
# ===============================
df.loc[df["Salary"] < 0, "Salary"] = 0
```

```python
# ===============================
# 4. HANDLE OUTLIERS
# ===============================
for col in ["Age", "Salary"]:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  if lower_bound < 0:
    lower_bound = 0
  # Cap lower outliers
  df.loc[df[col] < lower_bound, col] = lower_bound
  # Cap upper outliers
  df.loc[df[col] > upper_bound, col] = upper_bound
  print(f"{col}")
  print(f"Lower Bound = {lower_bound:.2f}")
  print(f"Upper Bound = {upper_bound:.2f}")
```

```python
# ===================================
# 5. HANDLE MISSING VALUES (NULLS)
# ===================================
print("\n--- 5. Handling Missing Values ---")
print("Null count before fixing:")
print(df.isnull().sum())
```

```python
# 5A: Numeric columns -> Fill with MEDIAN (better when outliers exist)
# Can also use MEAN here, but MEDIAN is safer
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Salary"].fillna(df["Salary"].median(), inplace=True)
```

```python
# 5B: Categorical column -> Fill with MODE (most frequent values)
df["Department_Region"].fillna(df["Department_Region"].mode(), inplace=True)
```

```python
# 5C: ALTERNATIVE METHOD - To REMOVE rows with nulls instead:
# df.dropna(inplace=True)   # Uncomment to use

print("\nNull count after fixing (should be all 0):")
print(df.isnull().sum())
```

```python
# ===================================
# 6. FINAL CHECK & SAVE
# ===================================
print("\n--- 6. Final Cleaned Data ---")
print("Final Shape: ", df.shape)
print("\nFirst 5 rows of cleaned data:")
print(df.head())
```

```python
# Save the clean version
df.to_csv('clean_data_sample.csv', index=False)
print("\nClean data saved as 'clean_data_sample.csv'")
```
